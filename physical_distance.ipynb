## Distancias físicas entre letras

```python
import numpy as np

def position_matrix():
    filas = ["qwertyuiop", "asdfghjklñ", "zxcvbnm"]

    dic = {}
    for i, fila in enumerate(filas):
        for j, letter in enumerate(fila):
            dic[letter] = [i,j]
    
    return dic

dic = position_matrix()

def euclidean(pos1: np.ndarray, pos2: np.ndarray):
    pos1 = np.array(pos1)
    pos2 = np.array(pos2)
    return np.sqrt(np.sum((pos1-pos2)**2))

def order_ascii(letter1, letter2=None):
    if letter2 is None and len(letter1) == 2:
        letter2 = letter1[1]
    elif letter2 is None:
        raise ValueError(["[ERROR]: Introduce solo un string de 2 letras o dos letras por separado."])
    return "".join(sorted(letter1+letter2))

def create_distances(matrix):
    distances = {}
    keys = list(matrix.keys())
    values = list(matrix.values())
    for i in range(len(keys)):
        for j in range(i+1, len(keys)):
            string = order_ascii(keys[i], keys[j])
            distances[string] = euclidean(values[i], values[j])
    
    return distances

distancias = create_distances(dic)
print(distancias)
```
